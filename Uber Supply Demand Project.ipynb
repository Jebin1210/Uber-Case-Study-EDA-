```python
lst = []
for i in data['Request_time']:
    if(int(i.split(":")[0])>0 and int(i.split(":")[0])<=4):
        lst.append("Late Night")
    elif(int(i.split(":")[0])>4 and int(i.split(":")[0])<=8):
        lst.append("Early Morning")
    elif(int(i.split(":")[0])>8 and int(i.split(":")[0])<=12):
        lst.append("Morning")
    elif(int(i.split(":")[0])>12 and int(i.split(":")[0])<=15):
        lst.append("Afternoon")
    elif(int(i.split(":")[0])>15 and int(i.split(":")[0])<=18):
        lst.append("Evening")
    elif(int(i.split(":")[0])>18 and int(i.split(":")[0])<=21):
        lst.append("Late Evening")
    else:
        lst.append("Night")
```

```python
data['Time'] = lst
data.drop(columns = {'Request timestamp'}, axis = 1, inplace = True)
```
